```python
import matplotlib.pyplot as plt
import csv

greedy_distance = []
greedy_score = []
steepest_distance = []
steepest_score = []

with open("../Results/Similarity/code253.csv",'r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        greedy_distance.append(float(row[0]))
        greedy_score.append(float(row[1]))
        steepest_distance.append(float(row[2]))
        steepest_score.append(float(row[3]))

plt.scatter(greedy_score, greedy_distance, marker='o', label="greedy")
plt.scatter(steepest_score, steepest_distance, marker='v', label="steepest")
plt.legend(loc='lower right')
plt.ylim(top = 255, bottom=230)
plt.xlabel('jakość rozwiązania')
plt.ylabel('odległość od najlepszego rozwiązania')
plt.savefig('initial_final.pdf')
plt.show()
```
