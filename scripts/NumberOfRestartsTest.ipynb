```python
import matplotlib.pyplot as plt
import csv

n = []
avg = []
mini = []

with open("../Results/Restart/code253.csv",'r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        n.append(float(row[0]))
        avg.append(float(row[1]))
        mini.append(float(row[2]))

plt.scatter(n, avg)
plt.xlabel('jakość rozwiązania początkowego')
plt.ylabel('jakość rozwiązania końcowego')
plt.savefig('initial_final.pdf')
plt.show()
```

```python
plt.scatter(n, mini)
plt.xlabel('jakość rozwiązania początkowego')
plt.ylabel('jakość rozwiązania końcowego')
plt.savefig('initial_final.pdf')
plt.show()
```
